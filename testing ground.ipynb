```python
def random_spin(s0, alpha=0.1):
    delta_s = (2 * np.random.random(3) - 1) * alpha
    s1 = s0 + delta_s
    return s1 / np.linalg.norm(s1)
```

```python
import mcsim
n = (5, 5)
s = mcsim.Spins(n=n)
s.randomise()

B = (0, 0, 0)
K = 0
u = (0, 0, 1)
J = 1
D = 0

system = mcsim.System(s=s, B=B, K=K, u=u, J=J, D=D)
#print(system.s.array)
#driver = mcsim.Driver()
def compare_energy(s):
    i = np.abs(int(((np.random.random()*2)-1)*s.array.shape[0]))
    j = np.abs(int(((np.random.random()*2)-1)*s.array.shape[1]))
    # Spin number {i},{j} has value {sij}
    old = np.copy(s.array)
    new = np.copy(s.array)
    new [i][j] = random_spin(s.array[i,j,:])
    # Create a copy of the spin that we currently have
    s_old = mcsim.Spins(n=n)
    s_old.array = old
    # Create a new Spin while changing the random spin
    s_new = mcsim.Spins(n=n)
    s_new.array = new
    # Now Creating the systems for the old and new spins
    old_system = mcsim.System(s=s_old, B=B, K=K, u=u, J=J, D=D)
    new_system = mcsim.System(s=s_new, B=B, K=K, u=u, J=J, D=D)
    # Now we compare the difference between the old and the new
    E_delta = new_system.energy()-old_system.energy()
    if E_delta<0:
        #Change Accepted
        return new_system
    else:
        #Change Rejected
        return old_system

print(system.energy())
for i in range(1000):
    system= compare_energy(system.s)
print(system.energy())
```
